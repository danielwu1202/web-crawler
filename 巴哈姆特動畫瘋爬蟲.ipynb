# 授權即將到期動畫

```python
import requests
from bs4 import BeautifulSoup
import time
import random


headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.45 Safari/537.36',
}
# 對"巴哈姆特動畫瘋"送出請求
r = requests.get('https://ani.gamer.com.tw/', headers=headers)
if r.status_code == 200:

    # 藉由 BeautifulSoup 套件將網頁原始碼使用 `html.parser` 解析器來解析
    soup = BeautifulSoup(r.text, 'html.parser')
    # 取得各個動畫主題區塊
    anime_theme = soup.select('.animate-theme-list > .theme-title-block')
    anime_theme_list = soup.select('.animate-theme-list > .theme-list-block')
    anime_theme = anime_theme[:-2]
    urls = []
    deathline = []
    
    if '授權到期作品' in anime_theme[-1].text.strip():
        expired_anime_name = anime_theme_list[-1].select('.theme-name')
        expired_anime_date = anime_theme_list[-1].select('.theme-time')
        expired_episode_amount = anime_theme_list[-1].select('.theme-number')
        expired_anime_img = anime_theme_list[-1].select('img.lazyload')
        expired_anime_href= anime_theme_list[-1].select('.theme-list-block > .theme-list-main')
        
        #取得授權到期日期
        for i in range(len(expired_anime_name)):
            urls.append('https://ani.gamer.com.tw/' + expired_anime_href[i].attrs['href'])
        
        for u in urls:
            expired_r = requests.get(u, headers = headers)

            expired_soup = BeautifulSoup(expired_r.text, 'html.parser')
    
            anime_deathline = expired_soup.select_one('.is-deathline')
            deathline.append(anime_deathline.text.strip())
            time.sleep(random.randint(1, 5))
    
    
        print(anime_theme[-1].text.strip())
        print('')
        for i in range(len(expired_anime_name)):
            print('動畫名稱：', expired_anime_name[i].text.strip())
            print('日期：', expired_anime_date[i].text.strip())
            print('總集數：', expired_episode_amount[i].text.strip())
            print(deathline[i])
            print('縮圖：', expired_anime_img[i].get('src'))
            print('網址:', 'https://ani.gamer.com.tw/' + expired_anime_href[i].attrs['href'])
            print('----------')
    else:
        print('目前沒有即將到期動畫')

else:
    print(f'請求失敗：{r.status_code}')
```

# 新上架動畫

```python
if r.status_code == 200:

    # 藉由 BeautifulSoup 套件將網頁原始碼使用 `html.parser` 解析器來解析
    soup = BeautifulSoup(r.text, 'html.parser')
    # 取得各個動畫主題區塊
    anime_theme = soup.select('.animate-theme-list > .theme-title-block')
    anime_theme_list = soup.select('.animate-theme-list > .theme-list-block')
    anime_theme = anime_theme[:-2]
    
    anime_name = anime_theme_list[0].select('.theme-name')
    anime_date = anime_theme_list[0].select('.theme-time')
    episode_amount = anime_theme_list[0].select('.theme-number')
    anime_img = anime_theme_list[0].select('img.lazyload')
    anime_href = anime_theme_list[0].select('.theme-list-block > .theme-list-main')
           
    print(anime_theme[0].text.strip())
    print('')
    for i in range(len(anime_name)):
        print('動畫名稱：', anime_name[i].text.strip())
        print('日期：', anime_date[i].text.strip())
        print('總集數：', episode_amount[i].text.strip())
        print('縮圖：', anime_img[i].get('src'))
        print('網址:', 'https://ani.gamer.com.tw/' + 
              anime_href[i].attrs['href'])
        print('----------')
            
else:
    print(f'請求失敗：{r.status_code}')
```

# 目前季番

```python
if r.status_code == 200:
    # 藉由 BeautifulSoup 套件將網頁原始碼使用 `html.parser` 解析器來解析
    soup = BeautifulSoup(r.text, 'html.parser')
    # 取得各個動畫元素區塊
    newanime_item = soup.select_one('.timeline-ver > .newanime-block')
    anime_items = newanime_item.select('.newanime-date-area:not(.premium-block)')

    print('季番更新')
    print('')
    # 依序針對每個動畫區塊擷取資料
    for anime_item in anime_items:
        anime_name = anime_item.select_one('.anime-name > p').text.strip()
        print(anime_name)  # 動畫名稱
        anime_episode = anime_item.select_one('.anime-episode').text.strip()
        print(anime_episode)  # 動畫集數
        anime_href = anime_item.select_one('a.anime-card-block').get('href')
        print('https://ani.gamer.com.tw/'+anime_href)  # 觀看連結

        # contents：將 tag 的子節點以列表的方式輸出
        anime_date = anime_item.select_one('.anime-date-info').contents[-1].string.strip()
        anime_time = anime_item.select_one('.anime-hours').text.strip()
        print(anime_date, anime_time)  # 日期與時間
        anime_img = anime_item.select_one('img.lazyload').get('src')
        print(anime_img)  # 動畫縮圖

        print('----------')
else:
    print(f'請求失敗：{r.status_code}')
```
